```python
import pandas as pd
```

```python
# save filepath to variable for easier access
melbourne_file_path = 'melb_data.csv'
# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path) 
# print a summary of the data in Melbourne data
melbourne_data.describe()
```

```python
#List Of Columns
melbourne_data.columns
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
y = melbourne_data.Price
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)
# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree
import numpy as np
import pandas as pd

# Sample dataset (House Prices)
data = pd.DataFrame({
    'Square_Feet': [800, 1000, 1200, 1500, 2000],
    'Bedrooms': [2, 2, 3, 3, 4],
    'Price': [100000, 120000, 150000, 180000, 250000]
})

# Features and target
X = data[['Square_Feet', 'Bedrooms']]
y = data['Price']

# Train a simple Decision Tree Regression model
tree_reg = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_reg.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(12, 6))
plot_tree(tree_reg, feature_names=['Square_Feet', 'Bedrooms'], filled=True, rounded=True)
plt.show()
```
